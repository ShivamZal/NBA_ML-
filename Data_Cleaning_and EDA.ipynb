```python
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])

# Extract year from 'GAME_DATE_EST'
games['YEAR'] = games['GAME_DATE_EST'].dt.year % 100

# Group by year and sum the number of home and away team wins
wins_per_year = games.groupby('YEAR')[['HOME_TEAM_WINS', 'AWAY_TEAM_WINS']].sum()

# Plotting the bar chart
years = wins_per_year.index
home_wins = wins_per_year['HOME_TEAM_WINS']
away_wins = wins_per_year['AWAY_TEAM_WINS']

```

```python
bar_width = 0.35
index = range(len(years))

fig, ax = plt.subplots()
bar1 = ax.bar(index, home_wins, bar_width, label='Home Wins')
bar2 = ax.bar([i + bar_width for i in index], away_wins, bar_width, label='Away Wins')

ax.set_xlabel('Year')
ax.set_ylabel('Number of Wins')
ax.set_title('Home and Away Wins per Year')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(years)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Convert 'GAME_DATE_EST' to datetime format
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])

# Extract year from 'GAME_DATE_EST'
games['YEAR'] = games['GAME_DATE_EST'].dt.year

# Drop rows with NA values in selected columns
games_cleaned = games.dropna(subset=['FG_PCT_home', 'FG3_PCT_home', 'FG_PCT_away', 'FG3_PCT_away'])

# Group by year and calculate the average of PTS_home and PTS_away
average_scores = games_cleaned.groupby('YEAR')[['PTS_home', 'PTS_away']].mean()

# Plot side by side bar plots for average scores
average_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores (PTS_home vs PTS_away) by Year')
plt.xlabel('Year')
plt.ylabel('Average Points')
plt.xticks(rotation=45)
plt.legend(['PTS_home', 'PTS_away'])
plt.show()

```

```python
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])

# Extract year from 'GAME_DATE_EST'
games['YEAR'] = games['GAME_DATE_EST'].dt.year

# Drop rows with NA values in selected columns
games_cleaned = games.dropna(subset=['FG_PCT_home', 'FG3_PCT_home', 'FG_PCT_away', 'FG3_PCT_away'])

# Calculate the mean for each year
mean_values = games_cleaned.groupby('YEAR')[['FG_PCT_home', 'FG3_PCT_home', 'FG_PCT_away', 'FG3_PCT_away']].mean()

# Plot line plots for mean values
plt.figure(figsize=(10, 6))

plt.plot(mean_values.index, mean_values['FG_PCT_home'], color='blue', label='FG_PCT_home')
plt.plot(mean_values.index, mean_values['FG3_PCT_home'], color='green', label='FG3_PCT_home')
plt.plot(mean_values.index, mean_values['FG_PCT_away'], color='red', label='FG_PCT_away')
plt.plot(mean_values.index, mean_values['FG3_PCT_away'], color='orange', label='FG3_PCT_away')

plt.title('Mean Field Goal Percentage and 3-Point Field Goal Percentage by Year')
plt.xlabel('Year')
plt.ylabel('Mean Percentage')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
